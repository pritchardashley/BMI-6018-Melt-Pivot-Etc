```python
import pandas as pd
import numpy as np
import zipfile
```

```python
#upload data
with zipfile.ZipFile('breast+cancer.zip', 'r') as zip_ref:
    zip_ref.extractall()
```

```python
#inport data

df = pd.read_csv('breast-cancer.data', header = None, sep = ',')

df.columns = [
    "recurrence", "age", "menopause", "tumor-size", "inv-nodes", 
    "node-caps", "deg-malig", "breast", "breast-quad", "irradiat"
]
df.head()
```

```python
#Melt
tumor_melt = ['recurrence', 'age', 'menopause', 'node-caps', 'deg-malig']

df_melt = df.melt(id_vars=tumor_melt, var_name='Tumor', value_name='value')

print(df_melt.head())
```

```python
#Pivot
df_pivot = df.pivot_table(index='age', columns='breast', values='deg-malig', aggfunc='mean')
df_pivot.head()
```

```python
#Aggregation

df_aggregation = df.groupby('age').agg({'tumor-size': 'count', 'deg-malig': 'mean'})
df_aggregation.head()
```

```python
#Iteration
for index, row in df.iterrows():
    print(row)
```

```python
# Groupby
df_group = df.groupby('recurrence')['tumor-size'].value_counts()
df_group.head()
```
